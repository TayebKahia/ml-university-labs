1. Why is data missing?

Data can be missing due to various reasons, such as:

Human error: Data entry mistakes or omissions.
Data collection issues: Problems during data acquisition (e.g., faulty sensors).

Privacy concerns: Respondents may deliberately omit sensitive information.

Data integration problems: When merging datasets, some fields may not be available in all datasets.

These causes of missing data may dictate how you handle the issue.

```python
samples_with_missing = data[data.isnull().any(axis=1)]
print(samples_with_missing)
```

```python
missing_features = data.columns[data.isnull().any()].tolist()
print("Features with missing values:", missing_features)
```

```python
samples_with_missing_features = data[data[missing_features].isnull().any(axis=1)]
print(samples_with_missing_features)
```

```python
missing_counts = data.isnull().sum()
print(missing_counts)
```

```python
data_no_missing_samples = data.dropna()
print(data_no_missing_samples)
```

```python
data_no_missing_columns = data.dropna(axis=1)
print(data_no_missing_columns)
```

Pros: This can simplify your dataset, especially if the columns with missing values are not important or if they contain a large percentage of missing data.

Cons: Dropping columns can lead to loss of valuable information, especially if the columns are important for your analysis. It is not recommended if the missing values are few and the features are crucial.

```python
# For continuous features, replace missing values with the mean
data_continuous = data.select_dtypes(include=['float64', 'int64'])
data_continuous_filled = data_continuous.fillna(data_continuous.mean())

# For categorical features, replace missing values with the mode
data_categorical = data.select_dtypes(include=['object','bool'])
data_categorical_filled = data_categorical.apply(lambda x: x.fillna(x.mode()[0]))

# Combine the two dataframes
data_imputed = pd.concat([data_continuous_filled, data_categorical_filled], axis=1)

# Print the number of missing values after imputation
print(data_imputed.isnull().sum())
```
